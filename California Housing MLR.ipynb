### MLR

```python
X = cali[['longitude','latitude','total_rooms','population','income','households','housing_age','ocean_proximity']]
y = cali['house_value']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.75, random_state = 42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class MostFrequentImputer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        self.most_frequent_ = pd.Series([X[c].value_counts().index[0] for c in X],index=X.columns)
        return self
    def transform(self, X, y=None):
        return X.fillna(self.most_frequent_)
```

```python
cat_pipeline = Pipeline([
    ("imputer", MostFrequentImputer()),
    ("cat_encode", OneHotEncoder()),
])
```

```python
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy='median')),
    ("std_scaler", StandardScaler()),
])
```

```python
from sklearn.compose import ColumnTransformer
num_attribs = ['longitude','latitude','total_rooms','population','income','households','housing_age']
cat_attribs = ['ocean_proximity']
full_pipeline = ColumnTransformer([
    ('num',num_pipeline,num_attribs),
    ("cat",cat_pipeline,cat_attribs)
])
```

```python
X_train_prep = full_pipeline.fit_transform(X_train)
X_train_prep.shape
```

```python
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train_prep)
lr = sm.OLS(y_train, X_train_sm).fit()
```

```python
lr.params
```

```python
lr.summary()
```

```python
print(lr.summary())
```

```python
X_test_prep = full_pipeline.fit_transform(X_test)
X_test_prep.shape
```

```python
X_test_sm = sm.add_constant(X_test_prep)
```

```python
y_pred = lr.predict(X_test_sm)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
